Ideologies and subcultures are partially overlapping, both in space (around the world) and in time (some are long gone, some are yet to come). One hallmark of a successful subculture is its willingness to learn from the others and to engage in "cultural appropriation" which to some sounds like an ethical breach. 

In the folklore of the settling of the New England territories, some of the earliest settlers received guidance and training, basic survival skills, from indigenous peoples. 

The New Englanders found these practices entirely apropos and so adopted them into their culture, even if indigenous peoples themselves were ultimately unappreciated by many of the colonist. 

Settlers such as John Woolman (a Friend) advocated dealing with indigenous peoples as equals and engaging in fair business practices and co-ventures with them.
