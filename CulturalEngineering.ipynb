# SECURING A CULTURE

Cultography is the tongue-in-cheek "Category Theory" of anthrology, a language for encoding enlightened nonsense about the various pattern languages used to rule the various roosts we call human settlements, which double as habitats and guinea pig testing zones for various ethnic protocols and practices.

Zoltar represents the homunculus with a viewpoint. His reality bubble is his currently operational mental construct through which he sees his world. Zoltar is likely a Bayesian but we'll need more interactions to be sure.

As long as Zoltar's bubble is secure, he's invested in it and his energy circulates in his belief system (model). If the model starts to crumble, releasing energy, Zoltar's freedom is to spend this energy in another way.

We could say that Zoltar finds the world puzzling, and in his attempt to "solve" the puzzle, he "puzzles" which could mean: he "actively infers". The jigsaw puzzle offers many use analogies, such as the "aha!" experience of finally finding that piece you knew had to be there. Crossword puzzles also offer "ahas!". Other times, you might sense you're forcing a piece that doesn't actually fit, or in a crossword you'll have provisional answers, written in pencil maybe, that you're ready to undo in the face of future updates.

The "free energy principle" is saying: once a puzzle is solved, it might as well "go unconscious" as there's nothing more to puzzle about. It "goes unconscious" in that sense of "taken for granted" and now part of the background we'll call "solved universe". We know how to tie our shows, to ride a bike, to send an email... these might have been puzzles at one point.

If a puzzle starts to fall apart, or was never completed in the first place, then you have "free energy" in proportion to what we could call the "entropy" of the challenge. One consciously begins "paying attention" i.e. reverting to "puzzling out" behavior, both in thought and in action. Sherlock Holmes is back on the job. The clues are pointing somewhere else now.

When an apparently solved puzzle is suddenly unsolved, that could be seen as akin to a bubble bursting. The amount of free energy available sharply increases, in an effort to either put humpty back together, or to abandon the humpty project, perhaps moving on to something else, like the dumpty project. Such a bubble bursting moment might be preceded by the gradual appearance of cracks, which alone are sufficient to raise the attention level.

Statisticians typically think in terms of probability, expectation, and cross-entropy. Expectation is like a probability on top of probability. How likely is this team to live up to its potential this time, versus last time?  The entropy is a measure of volatility (the same v as in VUCA). How eratic is this team, how unpredictable? We've all seen teams with the capacity to be brilliant, skewing their average to above the usual, yet giving investors little reason to expect such a performance at any given meet. The odds stay high.

If this all sounds like some kind of Casino Math, that's on purpose. We have a taxonomy of four maths, interacting, when carving it up in the Silicon Forest: casino, supermarket, martian and neolithic. The neolithic to martian vector is past to future, giving us the partially overlapping timelines approach to history we so treasure. The casino to supermarket axis might be drawn perpendicular to the past to future axis, as every age is confronted with the problems of risk, and of distributing a bonanza, harvest, payoff or dividend.

Ideologies and subcultures are partially overlapping, both in space (around the world) and in time (some are long gone, some are yet to come). One hallmark of a successful subculture is its willingness to learn from the others and to engage in "cultural appropriation" which to some sounds like an ethical breach. 

In the folklore of the settling of the New England territories, some of the earliest settlers received guidance and training, basic survival skills, from indigenous peoples. 

The New Englanders found these practices entirely apropos and so adopted them into their culture, even if indigenous peoples themselves were ultimately unappreciated by many of the colonist. 

Settlers such as John Woolman (a Friend) advocated dealing with indigenous peoples as equals and engaging in fair business practices and co-ventures with them.
