# Probability

The word "probability" occurs in a "word cloud" of related concepts, such as uncertainty, likelihood, random, determined, sample space, population, average, center of gravity, mass, density, along with proper names, such as Laplace, Gauss, Bayes and Cardano. 

Weaving throughout is the mathematics, the musical notation of STEM. Figuring out the odds when gambling, vs-a-vs various games of chance, got the ball rolling.

Probability will be described with "probability density functions" (PDFs) which, when integrated (the opposite of differentiated), become the "cumulative distrubution functions" (CDFs).

The most well known PDF is the Bell Curve, which owes its mathematical expression to Carl Freidrich Gauss. Ergo may we call it a Gaussian. We also call it a Normal Distribution.
$$
  X \sim \mathcal{N}(\mu,\,\sigma^{2})
$$
A Bell Curve may be characterized by two values: mean and variance. Variance describes the width of the bell, whereas the mean is where it peaks.

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi} } e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}
$$

```python
import pandas as pd
import numpy as np
from numpy import sqrt, pi, exp
from matplotlib import pyplot as plt
```

```python
x = np.arange(-9, 9.1, 0.1)
```

```python
sigma = 3 # np.std(x)
mu = 0    # np.mean(x)
```

```python
def f(x):
    return  (1/sigma*sqrt(2*pi)) * exp( (-1/2)*( (x - mu)/sigma )**2)
```

```python
F = np.vectorize(f)
```

```python
plotme = pd.DataFrame({"x": x, "y": F(x)})
plotme
```

```python
ax = plotme.plot(x='x', y='y', xlim=(-10, 10), ylim=(0, 1), grid=True)
plt.figtext(0.1, -0.05, "Fig 1: The Bell Curve")
plt.title("Normal Distribution")
ax.axvline(x = -2.5, color = 'black', linestyle=':')
ax.axvline(x =  2.5, color = 'black', linestyle=':');
```

The probability the some random variable x falls between -2.5 and 2.5 is the area bounded by the dotted lines and the curve between them.
